```python
import csv
import random
#import xlrd
import time
fname=input('输入文件：')
#data=csv.read('blogporn.csv','r')
#data = xlrd.open_workbook('blogporn.csv')
#fname='blogporn.csv'
f=open(fname,'r')
line1=f.readline()
date=line1.split('\t')[2].replace('\n','')
count=len(f.readlines())+1#前面readline()使其减少了一行
f.close()
print('文件\t'+fname+'\n共\t'+str(count)+'条数据!'+'\n数据日期\t'+str(date))
#outname=input('输出文件：')
outname=u'/Users/litian/PROGRAMMING/f-porndata/Blogporn'+date.replace('-','')+u'.csv'
print('输出路径：'+outname)
```
